####Functions and **if** Statements

```python
def has_student_passed(score):
  """
  This is a multiline comment.
  This function takes the score and returns PASS
  if the score is above 40
  """
  if score>=40: #Whenever if is used, we need to end with a colon
    print(f"You have passed with {score} marks!") #Gets executed if condition evaluates to TRUE
  else:
    print(f"You failed by {40 - score} marks:( \nBetter luck next time!") #Gets executed if condition evaluates to FALSE
  return #After running, returns to the top to repeat

has_student_passed(92)
```

####For loop

```python
for s in scores:#Takes list/array/collection/iterable and each time picks out one item and stores it in the variable s
  has_student_passed(s)
#print(f"s={s}, {has_student_passed(s)}") would return None as well since there is no return value
```

##Homework:
Write a function to calculate grades and return the grade

```python
def get_grade(score):
  """
  This function takes the score and returns the grade
  """
  if score>=40 and score<=50:
    return "E"
  elif score>50 and score<=60:
    return "D"
  elif score>=60 and score<=70:
    return "C"
  elif score>=70 and score<=80:
    return "B"
  elif score>80 and score<=90:
    return "A"
  elif score>=90 and score<=100:
    return "O"
  return "U"


get_grade(99)
#If no return at the end, anything outside the conditions gives no output
#return is needed for completeness
```

#### Slicing arrays

```python
food_prices = np.array([530, 293, 200, 501, 293, 706, 877, 291])
print(food_prices[1]) #prints the second element of the array because indexing starts from zero
print(food_prices[3:6]) #prints the fourth to sixth elements; the start index is inclusive, the end index is not
print(food_prices[3:]) #prints from the fourth element till the end of the array
print(food_prices[:6]) #prints from the start to the 6th (including) element
print(food_prices[:-1]) #leaves one element from the end of the list and prints until then
print(food_prices[-1:-3]) #doesn't work to reverse the list
print(food_prices[-3:-1]) #prints the last 2 elements
print(food_prices[::2]) #adds a step, prints every second element in the list
print(food_prices[::-1]) #prints the list in reverse
print(food_prices[5:3:-1])
```
